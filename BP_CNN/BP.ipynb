```python
import numpy as np
```

```python
# sigmoid激活函数
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

关于权值矩阵
* 第八个参数一行
[w85,w86.w87]
* 第九个参数一行
[w95,w96,w97]

* 输出层误差  
[δ8]  
[δ9]

85δ8 + 95δ9

关于更新权值  
第一层：3*4  
δ5x1+δ5x2+δ5x3+δ5x4  
δ6x1+δ6x2+δ6x3+δ6x4  
δ7x1+δ7x2+δ7x3+δ7x4  

第二层：2*3  
δ8a5+δ8a6+δ8a7  
δ9a5+δ9a6+δ9a7  

```python
# BP函数
def BP(input1,real,w1,w2):
    for s in range(100):
        # 前向传播（隐层a，输出层y）
        output1 = sigmoid(np.dot(w1, input1)) # 第一层输出，输入向量由4*1变成3*1
        output2 = sigmoid(np.dot(w2, output1)) # 第二层输出，输入向量由3*1变成2*1
        
        # 计算损失
        cost = np.square(real-output2)/2 # 计算Ed，目标函数：误差平方和
        
        # 计算误差
        delta2 = output2*(1-output2)*(real-output2) # 计算误差δ2，输出层误差
        delta1 = output1*(1-output1)*w2.T.dot(delta2) # 计算误差δ1，输出3*1
    
        # 更新权值，学习率为1，没有偏置项
        w1 = w1 + delta1.dot(input1.T)
        w2 = w2 + delta2.dot(output1.T)
        
        # 打印查看结果
        print("---round"+str(s)+"---")
        print("output1结果\n"+ str(output1))
        print("output1结果\n"+ str(output2))
        print("损失函数值\n"+ str(cost))
        print("\n")
        
```

BP的编写
* 输入层：四个节点
* 隐含层：三个节点
* 输出层：两个节点

```python
# 输入数据(四个特征输入：4*1)
input1 = np.array([[0.35], [0.9], [0.58], [0.78]])

# 第一层权值矩阵（初始随意赋值）
w1 = np.random.rand(3,4)
# print(w1)

# 第二层权值参数（初始随意赋值）
w2 = np.random.rand(2,3)

# 输出标签
real = np.array([[0.5],[0.7]])

# 调用BP函数
BP(input1,real,w1,w2)
```
